1)Countries population data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming the data is in a CSV file named 'loan_data.csv')
loan_data = pd.read_csv('/content/Loan_Default.csv')

# Visualization 1: Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=loan_data)
plt.title('Gender Distribution')
plt.show()
```

```python
# Visualization 2: Loan Approval Status
plt.figure(figsize=(6, 6))
loan_data['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Loan Approval Status')
plt.show()
```

```python
# Visualization 3: Applicant Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(loan_data['age'], bins=20, kde=True)
plt.title('Applicant Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python

# Visualization 4: Loan Amount vs. Applicant Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='loan_limit', data=loan_data)
plt.title('Loan Limit vs. Applicant Age')
plt.xlabel('Age')
plt.ylabel('Loan Limit')
plt.show()
```

```python
# Visualization 5: Credit Worthiness and Loan Approval
plt.figure(figsize=(6, 4))
sns.countplot(x='Credit_Worthiness', hue='Status', data=loan_data)
plt.title('Credit Worthiness vs. Loan Approval')
plt.xlabel('Credit Worthiness')
plt.ylabel('Count')
plt.show()
```

```python

# Visualization 6: Region Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Region', data=loan_data)
plt.title('Region Distribution')
plt.xlabel('Region')
plt.show()
```

2) Countries population

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
population_data = pd.read_csv('/content/2023_Countries by Population.csv')

# Visualization 1: Population Growth Over the Years
plt.figure(figsize=(10, 6))
plt.plot(population_data['pop1980'], label='1980')
plt.plot(population_data['pop2000'], label='2000')
plt.plot(population_data['pop2010'], label='2010')
plt.plot(population_data['pop2022'], label='2022')
plt.plot(population_data['pop2023'], label='2023')
plt.plot(population_data['pop2030'], label='2030')
plt.plot(population_data['pop2050'], label='2050')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Population Growth Over the Years')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python

# Visualization 2: Population Density Distribution
plt.figure(figsize=(8, 6))
sns.histplot(population_data['density'], bins=20, kde=True)
plt.title('Population Density Distribution')
plt.xlabel('Population Density')
plt.show()
```

```python
# Visualization 3: Top 10 Most Populous Countries in 2023
top_10_countries_2023 = population_data.sort_values(by='pop2023', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='pop2023', y='country', data=top_10_countries_2023)
plt.title('Top 10 Most Populous Countries in 2023')
plt.xlabel('Population in 2023')
plt.ylabel('Country')
plt.show()
```

```python
# Visualization 4: Population Growth Rate Distribution
plt.figure(figsize=(8, 6))
sns.histplot(population_data['growthRate'], bins=20, kde=True)
plt.title('Population Growth Rate Distribution')
plt.xlabel('Population Growth Rate')
plt.show()
```

```python

# Visualization 5: World Population Percentage by Country in 2023
population_data_clean = population_data.dropna(subset=['worldPercentage'])
plt.figure(figsize=(10, 8))
plt.pie(population_data_clean['worldPercentage'], labels=population_data_clean['country'], autopct='%1.1f%%', startangle=140)
plt.title('World Population Percentage by Country in 2023')
plt.show()
```

3) Stock Market

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
stock_data = pd.read_csv('/content/nifty_500.csv')

# Visualization 1: Industry-wise Distribution of Companies
plt.figure(figsize=(12, 6))
sns.countplot(y='Industry', data=stock_data, order=stock_data['Industry'].value_counts().index)
plt.title('Industry-wise Distribution of Companies')
plt.xlabel('Number of Companies')
plt.ylabel('Industry')
plt.show()
```

```python
# Visualization 2: Box Plot of 52 Week High and Low Prices
plt.figure(figsize=(8, 6))
sns.boxplot(x='52 Week High', y='Industry', data=stock_data, orient='h')
sns.boxplot(x='52 Week Low', y='Industry', data=stock_data, orient='h', color='red')
plt.title('52 Week High and Low Prices by Industry')
plt.xlabel('Price')
plt.ylabel('Industry')
plt.show()
```

```python
# Visualization 3: 365 Day Percentage Change Distribution
plt.figure(figsize=(8, 6))
sns.histplot(stock_data['365 Day Percentage Change'], bins=20, kde=True)
plt.title('365 Day Percentage Change Distribution')
plt.xlabel('Percentage Change')
plt.show()
```

```python

# Visualization 4: Volume vs. Value (Indian Rupee)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Share Volume', y='Value (Indian Rupee)', data=stock_data)
plt.title('Volume vs. Value (Indian Rupee)')
plt.xlabel('Share Volume')
plt.ylabel('Value (Indian Rupee)')
plt.show()
```

4) Flight Passenger

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the airline passenger dataset
airline_data = pd.read_csv('/content/test.csv')

# Visualization 1: Passenger Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=airline_data)
plt.title('Passenger Gender Distribution')
plt.show()
```

```python

# Visualization 2: Customer Type Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Customer Type', data=airline_data)
plt.title('Customer Type Distribution')
plt.show()
```

```python
# Visualization 3: Age Distribution of Passengers
plt.figure(figsize=(8, 6))
sns.histplot(airline_data['Age'], bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()

```

```python

# Visualization 4: Passenger Satisfaction by Class
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', hue='satisfaction', data=airline_data)
plt.title('Passenger Satisfaction by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Passenger Satisfaction by Type of Travel
plt.figure(figsize=(6, 4))
sns.countplot(x='Type of Travel', hue='satisfaction', data=airline_data)
plt.title('Passenger Satisfaction by Type of Travel')
plt.xlabel('Type of Travel')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 6: Departure Delay vs. Arrival Delay
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Departure Delay in Minutes', y='Arrival Delay in Minutes', data=airline_data)
plt.title('Departure Delay vs. Arrival Delay')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Arrival Delay (Minutes)')
plt.show()
```
